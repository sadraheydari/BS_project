```python
import pandas as pd
import numpy as np

from persiantools.jdatetime import JalaliDate
```

```python
data_df = pd.read_csv('../data/1k_refah_db.zip')
data_df
```

```python
def convert_to_datetime(x, limit):
    try:
        r = pd.to_datetime(x)
        if r > limit:
            return None
        return r
    except:
        return None
```

```python
data_df['BirthDate'] = data_df['BirthDate'].apply(
    lambda x: convert_to_datetime(x, JalaliDate(year=1399, month=6, day=1))
)
```

```python
na_df = data_df[['ParentID', 'BirthDate', 'postalcode']]\
                    .groupby('ParentID')\
                    .agg(lambda x: x.isna().sum())
```

```python
drop_hhid = set(na_df[na_df.sum(axis=1) > 0].index.values)
```

```python
data_df = data_df[[i not in drop_hhid for i in data_df['ParentID']]]
```

```python
data_df.to_csv('../data/1k_refah_db_clean.zip', index=False, compression={
    'method': 'zip', 'archive_name': '1k_refah_db_clean.csv'
})
```
